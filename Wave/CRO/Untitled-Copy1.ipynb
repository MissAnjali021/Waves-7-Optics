```python
#from numpy import sin,pi,linspace as np
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def res_var():
    data = pd.read_csv('var_res.csv')
    A_x = data['Ax(in V )']
    A_y = data['Ay(in V )']
    delta = data['delta( in rad)']
    t = np.linspace(-pi,pi)

    for i in range(0,5):
        x = A_x[i]*np.sin(100* t)
        y = A_y[i]*np.sin(100* t + delta[i])
        fig, ax = plt.subplots()
        plt.xticks(np.arange(-20, 20, 5))
        plt.yticks(np.arange(0.0015, 0.0015, 0.0005))
        ax.set_xlabel("x($V$)")
        ax.set_ylabel("y($V$)")
        plt.tight_layout()
        #ax.grid(True)
        plt.plot(x,y)
        plt.show()
        
def freq_var():
    data = pd.read_csv('var_freq.csv')
    A_x = data['Ax(in V )']
    A_y = data['Ay(in V )']
    delta = data['delta( in rad)']
    t = np.linspace(-pi/2,pi/2)

    for i in range(0,3):
        x = A_x[i]*np.sin(100* t)
        y = A_y[i]*np.sin(100* t + delta[i])
        fig, ax = plt.subplots()
        plt.xticks(np.arange(-20, 20, 5))
        plt.yticks(np.arange(0.0015, 0.0015, 0.0005))
        ax.set_xlabel("x($V$)")
        ax.set_ylabel("y($V$)")
        plt.tight_layout()
        #ax.grid(True)
        plt.plot(x,y)
        plt.show()
        
def delta_vs_freq():
    data = pd.read_csv('var_freq.csv')
    delta = data['delta( in rad)']
    freq = data['v(Hz)']
    
    plt.xscale('log')
    plt.tight_layout()
    plt.grid(True)
    plt.xlabel("$v$ $(Hz)$")
    plt.ylabel("$\Theta$ (rad)")
    plt.title("$\Theta$ vs $v$")
    plt.plot(freq,delta)
    plt.show()
    
def freq_not():
    data = pd.read_csv('var_freq.csv')
    v = data['v(Hz)']
    Ax = data['Ax(in V )']
    Ay = data['Ay(in V )']
    fig, ax = plt.subplots()
    ax.plot(v,Ax)
    ax.plot(v,Ay)
    ax.set_xlabel("$A_x(V)$")
    ax.set_ylabel("$A_y(V)$")
    plt.xticks(np.arange(12, 21, 1.0))
    plt.yticks(np.arange(0, 18, 2.5))
    #ax.annotate("(17,17)", (17,17),xytext=(17,17))
    plt.title("$A_x$ vs $A_y$")
    plt.grid(True)
    plt.show()
    

        
if __name__ == "__main__":
    res_var()
    freq_var()
    delta_vs_freq()
    freq_not()

```
