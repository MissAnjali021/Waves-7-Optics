```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def res_var():
    data = pd.read_csv('var_res.csv')
    A_x = data['Ax(in V )']
    A_y = data['Ay(in V )']
    delta = data['delta( in rad)']
    t = np.linspace(-np.pi,np.pi,1000)

    for i in range(0,5):
        x = A_x[i]*np.sin(2*100*np.pi*t)
        y = A_y[i]*np.sin(2*100*np.pi*t + delta[i])
        fig = plt.figure()
        ax = plt.axes(projection='3d')
        ax.set_xlabel("x($V$)")
        ax.set_ylabel("y($V$)")
        plt.tight_layout()
        ax.grid(True)
        plt.plot(x,y)
        plt.show()
        
def freq_var():
    data = pd.read_csv('var_freq.csv')
    A_x = data['Ax(in V )']
    A_y = data['Ay(in V )']
    delta = data['delta( in rad)']
    t = np.linspace(-np.pi,np.pi,1000)

    for i in range(0,3):
        x = A_x[i]*np.sin(2*100*np.pi*t)
        y = A_y[i]*np.sin(2*100*np.pi*t + delta[i])
        fig = plt.figure()
        ax = plt.axes(projection='3d')
        ax.set_xlabel("x($V$)")
        ax.set_ylabel("y($V$)")
        plt.tight_layout()
        ax.grid(True)
        plt.plot(x,y)
        plt.show()
        
def delta_vs_freq():
    data = pd.read_csv('var_freq.csv')
    delta = data['delta( in rad)']
    freq = data['v(Hz)']
    
    plt.xscale('log')
    plt.tight_layout()
    plt.grid(True)
    plt.xlabel("$v$ $(Hz)$")
    plt.ylabel("$\Theta$ (rad)")
    plt.title("$v$ vs $\Theta$")
    plt.plot(freq,delta)
    plt.show()
    


        
if __name__ == "__main__":
    res_var()
    freq_var()
    delta_vs_freq()
    
```
